```python
df=df.drop(columns=['Prevision j-1','Prevision j','Ech. physiques','Gaz - TAC', 'Gaz - Cogén', 'Gaz - CCG','Gaz', 'Ech. comm. Angleterre', 'Ech. comm. Espagne',
       'Ech. comm. Italie', 'Ech. comm. Suisse',
       'Ech. comm. Allemagne-Belgique', 'Fioul - TAC', 'Unnamed: 24',
       'Fioul - Autres',
       'Gaz - Autres', 'Hydraulique - Fil de l?eau + éclusée',
       'Hydraulique - Lacs', 'Hydraulique - STEP turbinage',
       'Bioénergies - Déchets', 'Bioénergies - Biomasse',
       'Bioénergies - Biogaz', ' Stockage batterie', 'Déstockage batterie',
       'Eolien terrestre', 'Eolien offshore'])
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.dropna().reset_index(drop=True)

print(df)
```

```python
X = df.drop(columns=['Solaire','Eolien'])  # Toutes les colonnes sauf celle de la cible
y_solaire = df['Solaire']
y_eolien = df['Eolien']
```

```python
X['Year'] = X['Date'].dt.year
X['Month'] = X['Date'].dt.month
X['Day'] = X['Date'].dt.day
X = X.drop('Date', axis=1)

if isinstance(X['Heures'].iloc[0], str) and ':' in X['Heures'].iloc[0]:
    X['Hour'] = X['Heures'].str.split(':').str[0].astype(int)
    X['Minute'] = X['Heures'].str.split(':').str[1].astype(int)
    X = X.drop('Heures', axis=1)
else:
    pass
```

```python

from sklearn.model_selection import train_test_split,GridSearchCV
X_train, X_test, y_train_solaire, y_test_solaire = train_test_split(X, y_solaire, test_size=0.2, random_state=42)
X_train, X_test, y_train_eolien, y_test_eolien = train_test_split(X, y_eolien, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 5, 7],
}
```

```python
grid_search = GridSearchCV(estimator=model,
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=3,
                           verbose=1,
                           n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train_solaire)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test_solaire, y_pred)
r2 = r2_score(y_test_solaire, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
from datetime import datetime, timedelta
current_time = datetime.now()
time_index = [current_time + timedelta(hours=i) for i in range(24)]

features_next_24h = pd.DataFrame({
    'Consommation': [0] * 24,
    'Fioul': [0] * 24,
    'Charbon': [0] * 24,
    'Nucléaire': [0] * 24,
    'Hydraulique': [0] * 24,
    'Pompage': [0] * 24,
    'Bioénergies': [0] * 24,
    'Taux de Co2': [0] * 24,
    'Year': [t.year for t in time_index],
    'Month': [t.month for t in time_index],
    'Day': [t.day for t in time_index],
    'Hour': [t.hour for t in time_index],
    'Minute': [t.minute for t in time_index]
})
features_next_24h = features_next_24h[['Consommation', 'Fioul', 'Charbon', 'Nucléaire', 'Hydraulique', 'Pompage', 'Bioénergies', 'Taux de Co2', 'Year', 'Month', 'Day', 'Hour', 'Minute']]

predictions = best_model.predict(features_next_24h)
results = pd.DataFrame({
    'timestamp': time_index,
    'predicted_solar_energy': predictions
})
print(results)
```
